```python
import random
import matplotlib.pyplot as plt

def insertion_sort(arr):
    comparisons = 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
            comparisons += 1
        arr[j + 1] = key
    return comparisons

def test_insertion_sort(input_size, num_instances):
    average_comparisons = 0
    for _ in range(num_instances):
        arr = [random.randint(1, 1000) for _ in range(input_size)]
        comparisons = insertion_sort(arr.copy())
        average_comparisons += comparisons
    return average_comparisons / num_instances

def plot_graph(input_sizes, avg_comparisons, n_log_n):
    plt.plot(input_sizes, avg_comparisons, label='Insertion Sort')
    plt.plot(input_sizes, n_log_n, label='n log n', linestyle='dashed')
    plt.xlabel('Input Size')
    plt.ylabel('Average Comparisons')
    plt.legend()
    plt.show()

input_sizes = list(range(30, 1001, 10))
num_instances = 10

avg_comparisons_insertion_sort = [test_insertion_sort(size, num_instances) for size in input_sizes]
n_log_n = [size * (size // 10) for size in input_sizes]  # Assuming n log n growth for comparison

plot_graph(input_sizes, avg_comparisons_insertion_sort, n_log_n)
```
