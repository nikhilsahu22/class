<h1>Create a sample database with age and income column and randomly generate 100 rows in each. Age should be less than 100 and income 20k to 1 lac and perform these funtions on it:</h1>
<h3><ol>
    <li> Standardisation/Normalisation</li>
    <li> Apply logarithmic transformation to the "income" feature</li>
    <li> Aggregate the data based on age and calculate the mean income for each age group</li>
    <li> Discretize the "Age" feature into three bins: Young, Adult and Elderly</li>
    <li> Binarizing "Age" feature where values above 30 become 1 and below or equal to 30 become 0</li>
    <li> Perform random sampling to select 10% of the data</li>
    </ol>

```python
import pandas as pd
import numpy as np

dict1={"Age":np.random.randint(15,100,size=100),
       "Salary":np.random.randint(20000,100000,size=100)}
df=pd.DataFrame(dict1)
df.head()
```

<h1>#1
    Standardisation/Normalisation
    


```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

std=StandardScaler()
nrm=MinMaxScaler()

df_standardised=pd.DataFrame(std.fit_transform(df), columns=["Age","Salary"])
df_normalised=pd.DataFrame(nrm.fit_transform(df), columns=["Age","Salary"])

print("Standardised :\n", df_standardised.head())
print("\nNormalised :\n", df_normalised.head())
```

<h1>#2 Apply logarithmic transformation to the "income" feature
    

```python
df_transformed=df.copy()

df_transformed["log(e) Salary"]=np.log(df_transformed["Salary"])
print(df_transformed.head())
```

<h1>#3 Aggregate the data based on age and calculate the mean income for each age group
    

```python
df_agg=df.copy()
df_agg=df_agg.groupby("Age").agg("mean")
df_agg
```

<h1>#4 Discretize the "Age" feature into three bins: Young, Adult and Elderly
    

```python
age_bins=[0,18,60,np.inf]
bin_labels=["Young","Adult","Elderly"]

df_binned=df.copy()
df_binned["Age Group"]=pd.cut(df_binned["Age"],bins=age_bins,labels=bin_labels)
df_binned
```

<h1>#5 Binarizing "Age" feature where values above 30 become 1 and below or equal to 30 become 0
    

```python
from sklearn.preprocessing import Binarizer
#binarisation
threshold=30

binarizer=Binarizer(threshold=threshold)

df_binarized = binarizer.fit_transform(df[["Age"]])

df_binarized = pd.DataFrame(df_binarized, columns=["Binarized_Age"])

df_binarized = pd.concat([df,df_binarized], axis=1)

print("Binarized Dataset:")
print(df_binarized.head())
```

<h1>#6 Perform random sampling to select 10% of the data

```python
df_resampled=df.copy()

df_resampled=df.sample(frac=0.1,random_state=1)
df_resampled.shape
df_resampled
```
